# Using the ACSF-Reader

````python
from ACSReader import Reader
````

Creating a Reader() class

````python
reader = Reader()
````

## Loading the data

````python
reader.download_data(version=2, save_dir='.', unzip=True, keep_zip=False)
````

This command will download the dataset from [here](https://icosys.ch/acs-f2), unzip it and remove the .zip file.

## Creating a DataFrame



To create a DataFrame the reader need the path to the .xml files in the downloaded folder. 

```create_dataframe()``` has to point to the root folder of the dataset.

The datapath can be set manually using ```set_file_paths()```

By default all features are used to create the DataFrame to change that use ```set_targets()```

````python
df, device_df = reader.create_dataframe('./ACS-F2/')
````

````python
df.head()
````

````python
device_df.head()
````

## Other useful things

### Creating the Intersession Protocol

Splitting the electrical Devices into train and test set by their recording sessions.

````python
df_train, df_test = reader.create_intersession_protocol(df)
````

````python
df_train.head()
````

### Creating a Signature DataFrame

Creating a DataFrame containing the signatures. Returning a DataFrame of shape $NxM$, where N is the number of signatures and M the features in the signature

````python
signature_df = reader.create_signature_dataset(df_train)
````
